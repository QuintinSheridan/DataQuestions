```python
def prep_data(df):
    '''Function that processes df and returns test and train features and labels
    Args:
        df (df) with data
    Returns:
    X_train, X_test, y_train, y_test (numpy array): train test datasets
    '''
    data_df = df.copy()
    
    # convert categorical columns
    categoricals = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod']
    
    for col in categoricals:
        dummies = pd.get_dummies(df[col])
        data_df = pd.concat([data_df, dummies], axis=1)
        data_df.drop(col, axis=1, inplace=True)
    
    # make label column binary
    data_df['Churn'] = 1*(data_df['Churn']=='Yes')
    print(data_df['Churn'].head())
    
    y = data_df['Churn'].values
    data_df.drop('Churn', axis=1, inplace=True)
    
    # normalize features
    data_df.drop('customerID', axis=1, inplace=True)
    x = data_df.values #returns a numpy array
    
    #print(x[:3])
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    
    # perform test train split
    X_train, X_test, Y_train, Y_test = train_test_split(x,y)
    
    return X_train, X_test, Y_train, Y_test
    
```

```python
def fit_evaluate_model():
    # get test train data
    X_train, X_test, Y_train, Y_test = prep_data(df)
    # instantiate model
    rfc = RandomForestClassifier()
    
    # train model
    rfc.fit(X_train, Y_train)
    
    # predict on test
    Y_pred = rfc.predict(X_test)
    
    # evaluate model
    report = classification_report(Y_test, Y_pred)
    print(report)
    
    return rfc, report
    
    
    
    
    
    
```

```python
rfc, report = fit_evaluate_model()
```
